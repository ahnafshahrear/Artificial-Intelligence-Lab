### Importing the Libraries

```python
import numpy as np  
import matplotlib.pyplot as plt
```

### Function: y = ax + b 

```python
x = np.arange(-10, 10, 0.5, dtype = np.float64)
a = 7
b = 4
y = a * x + b

plt.figure(figsize = (13, 13))
plt.title("Function: y = ax + b")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "o", label = "y = f(x)")
plt.grid()

for i, (xi, yi) in enumerate(zip(x, y)):
    plt.text(xi, yi, f"({round(xi, 1)}, {round(yi, 1)})", fontsize = 7, ha = "left", va = "top")

plt.legend()
plt.show()
```
