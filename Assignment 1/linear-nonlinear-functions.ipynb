#### Assignment 1: Write a Python program to display values of y when 'x' is an independent variable having values [-10, 10, step_size = 0.1] 
- **y = wx + b; where 'w' and 'b' are constants**
- **y = x^2**
- **y = 1 / (1 + e^(-x))**
- **y = (e^x - e^(-x)) / (e^x - e^(-x))**
- **y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))**
- **y = g(f(x)) where u = f(x) = wx + b and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))**
- **y = g3(f3(v)) where w = f3(v) = w3y1 + w4y2 + b & g3(w) = 1 / (1 + e^(-w)) | y1 = g1(f1(x)) where u1 = f1(x) = w1x + b1 & g1(u1) = 1 / (1 + e^(-u1)) | y2 = g2(f2(x)) where u2 = f2(x) = w2x + b2 & g2(u2) = 1/ (1 + e^(-u2))**

### Importing the Libraries

```python
import numpy as np  
import matplotlib.pyplot as plt
```

### Variables

```python
x = np.arange(-10, 10, 0.1, dtype = np.float64)
w = 7
b = 4
figure_size = (13, 13)
```

### Function : y = wx + b

```python
y = w * x + b

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = wx + b")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "", label = "y = f(x) = wx + b")
plt.grid()
plt.legend()
plt.show()
```

### Function: y = x^2

```python
y = x * x

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "", label = "y = f(x) = x^2")
plt.grid()
plt.legend()
plt.show()
```

### Function : y = 1 / (1 + e^(-x))

```python
y = 1 / (1 + np.exp(-x))

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = 1 / (1 + e^(-x))")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "", label = "y = f(x) = 1 / (1 + e^(-x))")
plt.grid()
plt.legend()
plt.show()
```

### Function : y = (e^x - e^(-x)) / (e^x - e^(-x))

```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) - np.exp(-x))

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = (e^x - e^(-x)) / (e^x - e^(-x))")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "", label = "y = f(x) = (e^x - e^(-x)) / (e^x - e^(-x))")
plt.grid()
plt.legend()
plt.show()
```

### Function : y = g(f(x)) where u = f(x) = wx + b & y = g(u) = 1/ (1 + e^(-u))

```python
u = w * x + b
y = 1 / (1 + np.exp(-u))

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = g(f(x))")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "", label = "y = g(f(x))")
plt.grid()
plt.legend()
plt.show()
```

### Function : y = g(f(x)) where u = f(x) = wx + b & g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
u = w * x + b
y = (np.exp(u) - np.exp(-u)) / (np.exp(u) + np.exp(-u))

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = g(f(x))")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "", label = "y = g(f(x))")
plt.grid()
plt.legend()
plt.show()
```

### Function : y = g3(f3(v)) where w = f3(v) = w3y1 + w4y2 + b & g3(w) = 1 / (1 + e^(-w)) | y1 = g1(f1(x)) where u1 = f1(x) = w1x + b1 & g1(u1) = 1 / (1 + e^(-u1)) | y2 = g2(f2(x)) where u2 = f2(x) = w2x + b2 & g2(u2) = 1/ (1 + e^(-u2))


```python
w1, b1 = 3, 8
u1 = w1 * x + b1
y1 = 1 / (1 + np.exp(-u1))

w2, b2 = 5, 1
u2 = w2 * x + b2
y2 = 1 / (1 + np.exp(-u2))

w3, w4 = 2, 9
f3 = w3 * y1 + w4 * y2 + b
y = 1 / (1 + np.exp(-f3))

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = g3(f3(x))")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "", label = "y = g3(f3(x))")
plt.grid()
plt.legend()
plt.show()
```
